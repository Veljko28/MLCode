```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3]
y = [0,2,4,6]

plt.figure(figsize=(10,5), dpi=300)

plt.plot(x,y, label='2x', color='#0f0', linewidth=2, marker='^', linestyle='--')
plt.title('Graph Name', fontdict={'fontname': 'Roboto', 'fontsize': 20})
plt.xlabel('X Axies Label')
plt.ylabel('Y Axies Label')

x2 = np.arange(0,5,0.5)
y2 = x2*x2


plt.plot(x2,y2,'b.--' ,label='x^2')

#plt.xticks([0, 1,2,3])
#plt.yticks([0,2,4,6])

plt.legend()

plt.savefig('mygraph.png', dpi=300) # saves graph as png

plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(6,4), dpi=200)
bars = plt.bar(labels,values)

patterns = ['O', '*', '/']
colors = ['r','g','b']

for i in range(len(bars)): 
    # bars[i].set_color(colors[i]) # set color
    bars[i].set_hatch(patterns[i]) # add patterns 

    
    
plt.title('Bar graph')

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(0,5), dpi=300)

plt.title('Gas Prices (USD/galon)', fontdict={'fontweight': 'bold', 'fontsize': 18})

# plot all countries
#for country in gas:
#    if country != 'Year':
#        plt.plot(gas.Year, gas[country], marker='.')

plt.plot(gas.Year, gas.USA, label='United States', 'b.-')
plt.plot(gas.Year, gas.Canada, label='Canada', 'r^--')
plt.plot(gas.Year, gas['South Korea'], label='South Korea', 'g.-')
plt.plot(gas.Year, gas.Australia, label='Australia', 'y.-')


plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollar')

plt.savefig('Gas_price_figure.png', dpi=300)

plt.legend()
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#0cafe5')

plt.xticks(bins)

plt.ylabel('Number of player')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]


labels = ['Left', 'Right']
colors = ['#0cafe5', 'red']

plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f%%')

plt.title('Preferred Foot of FIFA Players')

plt.show()
    
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa[fifa.Weight < 125].count()[0]
mid_light = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
mid_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light,mid_light,medium,mid_heavy, heavy]
labels = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]
explode = (.4,.2,0,0,.4) 

plt.title('Weight Distribution of FIFA Players (in lbs)')
plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)


plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'linewidth': 2})

for box in boxes['boxes']:
    box.set(color='#0cafe5', linewidth=2)
    box.set(facecolor='#0e0e0e')

plt.show()
```
