```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

```python
def get_z(x,y):
    return np.sin(np.sqrt(x**2+y**2))

fig_7 = plt.figure(figsize=(8,5), dpi=100)
axes_7 = fig_7.add_axes([0.1,0.1,0.9,0.9], projection='3d')
x_4 = np.linspace(-6,6,30)
y_4 = np.linspace(-6,6,30)

x_4,y_4 =  np.meshgrid(x_4,y_4)
z_4 = get_z(x_4,y_4)

axes_7.view_init(45,55)
#axes_7.contour3D(x_4,y_4,z_4,80,cmap='Blues')
#axes_7.plot_wireframe(x_4,y_4,z_4, color='b')
axes_7.plot_surface(x_4,y_4,z_4, rstride=1, cstride=1, cmap='Blues', edgecolor='None')

axes_7.set_xlabel('x')
axes_7.set_ylabel('y')
axes_7.set_zlabel('z')

plt.show()
```
